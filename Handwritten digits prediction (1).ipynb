```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, ConfusionMatrixDisplay
from tensorflow.keras.callbacks import ModelCheckpoint
import tkinter as tk
from tkinter import Canvas
from PIL import Image,ImageDraw,ImageOps


# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Preprocess the data
x_train = x_train.reshape((x_train.shape[0], 28, 28, 1)).astype('float32') / 255
x_test = x_test.reshape((x_test.shape[0], 28, 28, 1)).astype('float32') / 255

# Convert labels to one-hot encoding
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Function to build and compile the model with dropout
def build_and_compile_model():
    model = Sequential([
        Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)),
        MaxPooling2D(pool_size=(2, 2)),
        Conv2D(64, kernel_size=(3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.25),  # Dropout layer
        Dense(64, activation='relu'),
        Dropout(0.5),  # Dropout layer
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Initialize KFold
kf = KFold(n_splits=5, shuffle=True, random_state=42)
model_checkpoint = ModelCheckpoint('best_model.keras', save_best_only=True,monitor='val_accuracy',mode='max')

# Lists to store metrics
train_accuracy_list = []
val_accuracy_list = []

# Function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, title='Confusion Matrix'):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.arange(10))
    disp.plot(cmap=plt.cm.Blues, values_format='d', ax=plt.gca())
    plt.title(title)
    plt.show()

# Perform 5-fold cross-validation
for fold, (train_index, val_index) in enumerate(kf.split(x_train), 1):
    print(f"Fold {fold}")
    x_train_cv, x_val_cv = x_train[train_index], x_train[val_index]
    y_train_cv, y_val_cv = y_train[train_index], y_train[val_index]

    # Build and compile the model
    model = build_and_compile_model()

    # Train the model
    history = model.fit(
        x_train_cv, y_train_cv,
        validation_data=(x_val_cv, y_val_cv),
        epochs=10,
        verbose=2,
        callbacks=[model_checkpoint]
    )

    # Record training and validation accuracy
    train_accuracy = history.history['accuracy'][-1]  # Last epoch accuracy
    val_accuracy = history.history['val_accuracy'][-1]  # Last epoch validation accuracy

    train_accuracy_list.append(train_accuracy)
    val_accuracy_list.append(val_accuracy)

    # Evaluate on validation set
    val_predictions = np.argmax(model.predict(x_val_cv), axis=1)
    val_labels = np.argmax(y_val_cv, axis=1)

    print(f'Training Accuracy: {train_accuracy:.4f}')
    print(f'Validation Accuracy: {val_accuracy:.4f}')


    # Plot confusion matrix for validation set
    plot_confusion_matrix(val_labels, val_predictions, title=f'Fold {fold} Validation Confusion Matrix')

# Average metrics
print(f'Average Training Accuracy: {np.mean(train_accuracy_list):.4f}')
print(f'Average Validation Accuracy: {np.mean(val_accuracy_list):.4f}')

# Load the best model
model = tf.keras.models.load_model('best_model.keras')

# Evaluate the model on the test set
test_predictions = model.predict(x_test)
test_predictions = np.argmax(test_predictions, axis=1)
y_test_labels = np.argmax(y_test, axis=1)

test_accuracy = accuracy_score(y_test_labels, test_predictions)
test_precision = precision_score(y_test_labels, test_predictions, average='weighted')

print(f'Test Accuracy: {test_accuracy:.4f}')
print(f'Test Precision: {test_precision:.4f}')


# Plot confusion matrix for test set
plot_confusion_matrix(y_test_labels, test_predictions, title='Test Set Confusion Matrix')



```

```python
# Import necessary libraries
import tkinter as tk
from tkinter import *
import numpy as np
import tensorflow as tf
from PIL import Image, ImageOps

# Load the trained model
model = tf.keras.models.load_model('best_model.keras')

# Define the main window
root = Tk()
root.resizable(0, 0)
root.title("Handwritten Digit Recognition")

# Initialize drawing parameters
lastx, lasty = None, None
image1 = Image.new("L", (200, 200), (255))
draw = ImageDraw.Draw(image1)

def clear_widget():
    global image1, draw
    cv.delete("all")
    image1 = Image.new("L", (200, 200), (255))
    draw = ImageDraw.Draw(image1)

def paint(event):
    global lastx, lasty
    x, y = event.x, event.y
    if lastx and lasty:
        cv.create_line((lastx, lasty, x, y), width=8, fill='black', capstyle=ROUND, smooth=TRUE, splinesteps=36)
        draw.line((lastx, lasty, x, y), fill=0, width=8)
    lastx, lasty = x, y

def reset(event):
    global lastx, lasty
    lastx, lasty = None, None

def predict_digit():
    image1_resized = image1.resize((28, 28))
    image1_inverted = ImageOps.invert(image1_resized)
    image1_array = np.array(image1_inverted).astype('float32') / 255
    image1_array = np.expand_dims(image1_array, axis=[0, -1])
    
    prediction = model.predict(image1_array)
    digit = np.argmax(prediction)
    result_label.config(text=str(digit))

# Create the Canvas widget
cv = Canvas(root, width=200, height=200, bg='white')
cv.grid(row=0, column=0, pady=2, sticky=W, columnspan=2)

cv.bind("<B1-Motion>", paint)
cv.bind("<ButtonRelease-1>", reset)

# Create the Buttons and Label
button_clear = Button(root, text="Clear", command=clear_widget)
button_clear.grid(row=1, column=0, pady=2)

button_predict = Button(root, text="Predict", command=predict_digit)
button_predict.grid(row=1, column=1, pady=2)

result_label = Label(root, text="", font=("Helvetica", 24))
result_label.grid(row=2, column=0, pady=2, columnspan=2)

# Start the Tkinter main loop
root.mainloop()
```
